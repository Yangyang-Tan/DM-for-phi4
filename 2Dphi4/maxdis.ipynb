```python
import h5py
import numpy as np
from scipy.spatial.distance import pdist

# Load data
data_path = 'trainingdata/cfgs_wolff_fahmc_k=0.26_l=0.022_8^2.jld2'
with h5py.File(data_path, 'r') as f:
    cfgs = np.array(f['cfgs'])

print('=== 原始数据 ===')
print(f'Shape: {cfgs.shape}')
print(f'Range: [{cfgs.min():.4f}, {cfgs.max():.4f}]')

# Normalize (same as FieldDataModule)
cfgs_min = cfgs.min()
cfgs_max = cfgs.max()
cfgs_norm = ((cfgs - cfgs_min) / (cfgs_max - cfgs_min) - 0.5) * 2

print('\\n=== 归一化后数据 (与 FieldDataModule 一致) ===')
print(f'Range: [{cfgs_norm.min():.4f}, {cfgs_norm.max():.4f}]')

# Compute distances on normalized data
n_samples = 2000
indices = np.random.choice(len(cfgs), n_samples, replace=False)
samples_norm = cfgs_norm[indices].reshape(n_samples, -1).astype(np.float32)

print(f'\\n=== 归一化数据的配对距离 ({n_samples} samples) ===')
distances = pdist(samples_norm, metric='euclidean')
print(f'Maximum: {distances.max():.4f}')
print(f'Mean:    {distances.mean():.4f}')
print(f'Min:     {distances.min():.4f}')

print(f'\\n理论最大值 (64x64, [-1,1]): sqrt(64*64*4) = {np.sqrt(128*128*4):.4f}')
```
